```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from pickle import dump, load
import time


data1 = pd.read_pickle('Roaster_training.pkl')


data = data1[0:500]


f, (a0, a1, a2, a3, a4, a5) = plt.subplots(6, 1, figsize=(10,10), gridspec_kw={'height_ratios': [3, 3, 3, 3, 3, 3]})



# plt.subplot(14, 1, 1)
# f.suptitle('Roaster Training', fontsize=20)
a0.plot(data["Ore_amps"], 'c-', label='Ore Feed')
a0.legend(loc=2, fontsize=12)
a0.set_ylabel('Amp', fontsize=16)
a0.tick_params(axis='x', labelsize=16)




# plt.subplot(14, 1, 2)
a1.plot(data["Sulfur_tph"], 'c-', label='Sulfur Prill')
a1.legend(loc=2, fontsize=12)
a1.set_ylabel('T/H', fontsize=16)
a1.tick_params(axis='x', labelsize=16)





# plt.subplot(14, 1, 3)
a2.plot(data["O2_scfm"], 'c-', label='O2')
a2.legend(loc=2, fontsize=12)
a2.set_ylabel('SCFM', fontsize=16)
a2.tick_params(axis='x', labelsize=16)



# plt.subplot(14, 1, 4)
a3.plot(data["Carbon_in"], 'm-', label='Carbon content in Ore')
a3.legend(loc=2, fontsize=12)
a3.set_ylabel('wt%', fontsize=16)
a3.tick_params(axis='x', labelsize=16)



# plt.subplot(14, 1, 5)
a4.plot(data["Sulf_in"], 'm-', label='Iron Sulfide content in Ore')
a4.legend(loc=2, fontsize=12)
a4.set_ylabel('wt%', fontsize=16)
a4.tick_params(axis='x', labelsize=16)




# plt.subplot(14, 1, 6)
a5.plot(data["CO3_in"], 'm-', label='Carbonate content in Ore')
a5.legend(loc=2, fontsize=12)
a5.set_ylabel('wt%', fontsize=16)
a5.tick_params(axis='x', labelsize=16)





a5.set_xlabel('Time (min)', fontsize=16)



plt.tight_layout()




plt.savefig('Roaster_training_data.eps', format='eps')
plt.savefig('Roaster_training_data.png')

plt.show()
```
