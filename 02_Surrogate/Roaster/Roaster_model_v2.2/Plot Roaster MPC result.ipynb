```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from pickle import dump, load
import time


data = pd.read_pickle('Roaster_MPC_result.pkl')


#last T1 T2 data fix
data["T1"][120] = 1023.969209
data["T2"][120] = 1037.521215


f, (a0, a1, a2, a3) = plt.subplots(4, 1, figsize=(10,8), gridspec_kw={'height_ratios': [3, 3, 3, 3]})



# plt.subplot(4, 1, 1)
a0.plot(data["T1"], 'r-', linewidth=2, label='T1')
a0.step(data["sp1"], 'b--', linewidth=2, label='SP1')
a0.legend(loc=2, fontsize=12)
a0.set_ylabel('T1 (F)', fontsize=16)
# f.suptitle('Roaster MPC result', fontsize=20)
a0.tick_params(axis='x', labelsize=16)




# plt.subplot(4, 1, 2)
a1.plot(data["T2"], 'r-', linewidth = 2, label='T2')
a1.step(data["sp2"], 'b--', linewidth=2, label='SP2')
a1.set_ylabel('T2 (F)', fontsize=16)
a1.legend(loc=2, fontsize=12)
a1.tick_params(axis='x', labelsize=16)





# plt.subplot(4, 1, 3)
a2.step(data["Ore"], 'm-', linewidth=2, label='Ore')
a2.set_ylabel('Ore (Amp)', fontsize=16)
a2.legend(loc=2, fontsize=12)
a2.tick_params(axis='x', labelsize=16)





# plt.subplot(4, 1, 4)
a3.step(data["Sulfur_prill"], 'c-', linewidth=2,  label='Sulfur_prill')
a3.set_ylabel('Sulfur prill (TPH)', fontsize=16)
a3.legend(loc=2, fontsize=12)
a3.tick_params(axis='x', labelsize=16)
a3.set_xlabel('Time (min)', fontsize=16)

plt.tight_layout()




plt.savefig('Roaster_MPC_result.eps', format='eps')
plt.savefig('Roaster_MPC_result.png')

plt.show()
```
